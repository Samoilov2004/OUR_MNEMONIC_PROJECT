```python
from sklearn.linear_model import LinearRegression

def rand_change(arr, procent):
    procent /= 100
    random_vals = np.random.uniform(-procent, procent, size=arr.shape)
    arr_modified = arr + np.abs(arr) * random_vals
    return arr_modified

def new_arrs(x1, y1):
    mask = (x1 > -100) & (x1 < 200)
    x1n = x1[mask]
    y1n = rand_change(y1[mask], 5)

    mask = (x1 > 200)
    x1r = x1[mask]
    y1r = y1[mask]

    x1l = x1n - 300
    y1l = rand_change(y1n, 5)

    x1 = np.hstack([x1l, x1n, x1r])
    y1 = np.hstack([y1l, y1n, y1r])

    return x1, y1

def approx(x1, y1):
    mask = (x1 < 200)
    x1n = x1[mask]
    y1n = y1[mask]
    
    model = LinearRegression()
    model.fit(x1n, y1n)

    predict_arr = [-390, 200].reshape(1, 2)
    return predict_arr, model.predict(predict_arr)

def linear_regression_plot(x, y):
    # Фильтруем массивы
    mask = x < 200
    x_filtered = x[mask].reshape(-1, 1)
    y_filtered = y[mask]

    # Обучаем модель линейной регрессии
    model = LinearRegression()
    model.fit(x_filtered, y_filtered)

    # Получаем коэффициенты регрессионной прямой
    slope = model.coef_[0]
    intercept = model.intercept_

    # Выводим уравнение регрессионной прямой
    print(f"Регрессионная прямая: y = {slope:.2f}x + {intercept:.2f}")

    plt.plot(x_filtered, np.zeros_like(x_filtered) + intercept, color="black")

def b_func(x, y):
    # Фильтруем массивы
    mask = x < 200
    x_filtered = x[mask].reshape(-1, 1)
    y_filtered = y[mask]

    # Обучаем модель линейной регрессии
    model = LinearRegression()
    model.fit(x_filtered, y_filtered)

    # Получаем коэффициенты регрессионной прямой
    slope = model.coef_[0]
    intercept = model.intercept_
    return intercept
    
```
