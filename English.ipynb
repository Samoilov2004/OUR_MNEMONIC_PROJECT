```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# loading dataset "Krebs cycle"
iris = load_iris()

# Convert dataset to pandas DataFrame
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
iris_df['target_names'] = iris_df['target'].map(dict(zip(iris.target_names, iris.target_names)))
```

```python
iris_df.head()
```

```python
y = iris_df['sepal length (cm)'] / 8
x = iris_df['petal length (cm)']

shifts = np.random.choice(np.arange(-0.5, 1, 0.01), size=len(x))
x = x + shifts
x = np.where(x > 3, x * 1.4, x)

x /= x.max()
```

```python
# Paint our plot
colors = ['red'] * len(x)
for i in range(len(x)):
    if np.random.rand() < x[i]:
        colors[i] = 'green'

plt.figure(figsize=(10, 6))
plt.scatter(x, y, c=colors)
plt.xlabel('$Presicion$')
plt.ylabel('$Recall$')
plt.show()
```

```python
from sklearn.cluster import KMeans

plt.figure(figsize=(10, 6))

# Combine x and y into one matrix
data = np.column_stack((x, y))

# Train the KMeans model
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(data)

# Get the predicted cluster labels
labels = kmeans.labels_

# Visualize the clusters
plt.scatter(x, y, c=labels, cmap='viridis')

# Set colors for the clusters
colors = ['green', 'red']
for i in range(2):
    plt.scatter(x[labels == i], y[labels == i], c=colors[i])

plt.show()
```
